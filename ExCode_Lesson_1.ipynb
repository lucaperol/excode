# Datatype

  - ## There are different catagories of data. **Known as dataypes.**

  -  **Integers** are whole numbers e.g. 42, 10, -5

  -  **Floats** are numbers with a decimal point e.g 5.45, 1.25, -10.0

  -  **Boolean** values are either true or false e.g. False, True

  -  **String** values are lines of text e.g. "This is a string"
  

```python
no_students = 37              #Integer variable
percentage = 0.56424          #Float variable
answer = False                #Boolean variable
another_answer = True         #Boolean variable
greeting = "Good Evening"     #String Variable

print("The integer value:", no_students)
print("The float value:", percentage)
print("The false boolean value:", answer)
print("The true boolean value:", another_answer)
print("The string value: " + greeting)    #Note how printing floats and any other variable types differ
```

# Variables

-  Variables are simply **containers** that hold (data) values. 
-  These containers can be accessed to interact with the values within.

-  Variables **must be one word** and **cannnot use keywords** (function names) such as **print**. 
-  You can use  **_**  (underscore) in variable names to represent spaces. Aka snake_casing

```python
# Further usage of the print() function &
# using variable in the print() function

university_name = "Exeter"  # Creating a variable, make it contain "Exeter"

print("I attend " + university_name + " University")  # Putting a variable inside a print statement

print("I am changing my University...")



university_name = "Bath"   # Editing/over-writing a variable

print("I now attend " + university_name + " University")
#                  ^                       ^
# manuel spacing in the sentence before & after a variable 

# Q: do you notice a difference way of joining sentences with a variable? (When it comes to displaying the same stlye of output)
print("I changed my university to",  university_name , "University" )
# Ans:                      comma ^ provide auto spacing between a sentence in the display
```

# Manipulating Variables

All variable types can be manipulated in some way or another using operators or mathematical symbols.


```python
first_word = "Hello "
second_word = "World!"

# Addition of strings
sentence = first_word + second_word     

print("sentence =", sentence)
print("sentence =" + sentence)
print("sentence =" + ' ' + sentence)
```

```python
# Multiplication of strings

repeat = 3 # create an integer variable

print("sentence =" + ' ' + sentence * repeat) 
```

```python
# you can assign math expression into a variable, 
# even with a variable in it.
# this can be useful when you want to make variables' value depends on others

first_no = 52 + 3               #Addition
second_no = first_no * 2        #Multiplication
third_no = second_no / 4        #Division
fourth = 55.5
fourth = fourth - 2             #It is possible to use operators on existing variables.

fifth = (55 + 56.7 + 76.5 + 88) / 5   #Multiple operators can be used at once.

print("first_no =", first_no)
print("second_no =", second_no)
print("third_no =", third_no)
print("fourth =", fourth)
print("fifth =", fifth)

# note: ^This also how you put both integer & string variables together
# but works with only with comma
```

**What happens when you run a python script (a python (.py) file):**

### Python code -> Translated -> Machine


1.   The interpreter checks the code line by line for errors, it stop if it encounters an error.
2.   If there are no errors the interpreter translates python into a byte language. 
3.   The byte code is then executed on a virtual machine and the program is run.
